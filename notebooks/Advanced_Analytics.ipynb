```python
import pandas as pd
import numpy as np


nav=pd.read_csv(
"../data/processed/clean_nav_history.csv"
)


nav["return"]=(
nav["nav"].pct_change()
)


VaR=np.percentile(
nav["return"].dropna(),
5
)


print("95% VaR:",VaR)
```

```python
score=pd.read_csv(
"../data/processed/fund_scorecard.csv"
)


score["recommendation_score"]=(
score.select_dtypes("number")
.mean(axis=1)
)


score.sort_values(
"recommendation_score",
ascending=False
).head(5)
```
