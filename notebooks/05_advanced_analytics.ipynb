```python
import pandas as pd
import numpy as np


nav = pd.read_csv(
"../data/processed/clean_nav_history.csv"
)


nav["date"] = pd.to_datetime(nav["date"])


nav.head()
```

```python
nav["daily_return"] = (
    nav.groupby("scheme_code")
    ["nav"]
    .pct_change()
)


nav.head()
```

```python
VaR = (
    nav["daily_return"]
    .dropna()
    .quantile(0.05)
)


print("95% Value at Risk:", VaR)
```

```python
risk = (
    nav.groupby("scheme_code")
    ["daily_return"]
    .agg(
        volatility="std"
    )
)


risk.sort_values(
"volatility"
).head()
```

```python
score = pd.read_csv(
"../data/processed/fund_scorecard.csv"
)


score["recommendation_score"] = (
    score.select_dtypes("number")
    .mean(axis=1)
)


recommendation = score.sort_values(
"recommendation_score",
ascending=False
)


recommendation.head()
```

```python
recommendation.to_csv(
"../data/processed/fund_recommendation.csv",
index=False
)

print("Recommendation file created")
```
